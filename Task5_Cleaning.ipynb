## Dataset: Titanic Dataset
This notebook demonstrates basic data cleaning using pandas.

```

import pandas as pd
import numpy as np
```

### Load Dataset

```

df = pd.read_csv("Titanic-Dataset.csv")
df.head()
```

### Inspect Dataset

```

df.info()
df.isnull().sum()
```

### Handle Missing Values

```

df["Age"] = df["Age"].fillna(df["Age"].median())
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
df.drop(columns=["Cabin"], inplace=True)
```

### Remove Duplicates

```

df.drop_duplicates(inplace=True)
```

### Feature Engineering

```

df["Age_Group"] = pd.cut(
    df["Age"],
    bins=[0,12,20,40,60,100],
    labels=["Child","Teen","Adult","Middle-Aged","Senior"]
)
```

### Save Cleaned Dataset

```

df.to_csv("cleaned_data.csv", index=False)
```


## Data Cleaning Notes
1. Loaded and inspected the dataset using pandas.
2. Filled missing numeric and categorical values using median and mode.
3. Removed irrelevant columns and duplicate records.
4. Converted data types and created Age_Group feature.
5. Exported cleaned dataset for further analysis.
